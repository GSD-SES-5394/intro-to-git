# Simple Data Visualization Example

This notebook demonstrates how to download simple datasets from github using `httr` and visualize them.

`httr` is a library for sending HTTP requests. Github exposes their raw files `raw.githubusercontent.com` domain. Which makes it easy to download dataset files.

```r
library(httr)
```

this particular data set is from CSSEGIS COVID-19 data set from Johns Hopkins University. It is a well formatted csv file in a github repo, which makes our job easy.

```r
url <- "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/04-30-2020.csv"
response <- GET(url)
text <- content(r, 'text')
table <- read.table(text=text, sep=",", header=TRUE)
```

There are too many states and provinces to put in a single bar graph so we're just going to show the top 25. State names are also quite long, so

```r
dtable <- table[order(-table$Deaths),][1:25,]
op <- par(mar=c(8,4,4,2)) # adjust the bottom margin to fit the state names
barplot(dtable$Deaths, names.arg=dtable$Province_State, main="COVID-19 Deaths", las=2)
rm(op)
```
